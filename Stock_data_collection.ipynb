```python
import requests
import pprint
import pandas as pd
```

```python
# Function to fetch stock data from Alpha Vantage
def fetch_stock_data(symbol):
    api_key = '8VIUS7A7H07O4J4T'  # Replace with your Alpha Vantage API key
    base_url = 'https://www.alphavantage.co/query'

    # Make an API request to retrieve stock data
    params = {
        'function': 'OVERVIEW',
        'symbol': symbol,
        'apikey': api_key
    }
    response = requests.get(base_url, params=params)
    data = response.json()

    return data
```

```python
# Fetch stock data for each stock and store in a list
def get_stock_data_dict_list(stocks):
    stock_data_list = []
    for stock in stocks:
        stock_data = fetch_stock_data(stock)
        stock_data_list.append(stock_data)
    return stock_data_list
```

```python
# Choose ten stocks randomly (example list)
stocks = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'FB', 'TSLA', 'JPM', 'V', 'NVDA', 'MA']
```

```python
stock_data_list = get_stock_data_dict_list(stocks)
df_stock_data_list = pd.DataFrame(stock_data_list)
df_stock_data_list = df_stock_data_list.dropna(subset=['Symbol'])
```
